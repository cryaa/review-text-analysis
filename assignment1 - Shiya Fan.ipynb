```python
import nltk
from nltk import FreqDist
from nltk.corpus import stopwords
import pandas as pd
```

```python
import csv
with open('fashion data.txt', "w") as my_output_file:
    with open('fashion data.csv', "r") as my_input_file:
        [ my_output_file.write(" ".join(row)+'\n') for row in csv.reader(my_input_file)]
    my_output_file.close()
```

```python
f = open('fashion data.txt','r')
raw = f.read()
raw = raw.replace('\n',' ')
```

```python
raw = raw.replace('\n',' ')
```

```python
tokens = nltk.word_tokenize(raw)
```

```python
print(tokens)
```

```python
#change all tokens into lower case 
words1 = [w.lower() for w in tokens]   #list comprehension 

#only keep text words, no numbers 
words2 = [w for w in words1 if w.isalpha()]

#encode the raw token list by utf-8
words3 = [w.encode('utf-8') for w in words2]
```

```python
#generate a frequency dictionary for all tokens 
freq = FreqDist(words2)

#sort the frequency list in descending order
sorted_freq = sorted(freq.items(),key = lambda k:k[1], reverse = True)
sorted_freq
```

```python
freq.plot(30)
```

```python
mystopwords = stopwords.words('english')
words = [w.lower() for w in tokens if w.isalpha() if w.lower()not in mystopwords]
```

```python
#Use Porter Stemmer 
porter = nltk.PorterStemmer()
stem1 = [porter.stem(w) for w in words]

#Encode with utf-8
#stem1 = [w for w in stem1]

#Get the frequency distribution 
freq1 = FreqDist(stem1)
#Sort the result
sorted_freq1 = sorted(freq1.items(),key = lambda k: k[1], reverse = True)
```

```python
sorted_freq1
```

```python
freq1.plot(30)
```

```python
#Use the Lancaster Stemmer
lancaster = nltk.LancasterStemmer()
stem2 = [lancaster.stem(w) for w in words]
#stem2 = [w for w in stem2]
freq2 = FreqDist(stem2)
sorted_freq2 = sorted(freq2.items(),key = lambda k: k[1], reverse = True)
```

```python
sorted_freq2
```

```python
freq2.plot(30)
```

```python
POS_tags = nltk.pos_tag(tokens) #use unprocessed 'tokens', not 'words'

#Generate a list of POS tags
POS_tag_list = [(word,tag) for (word,tag) in POS_tags if tag.startswith('N')]
```

```python
#Generate a frequency distribution of all the POS tags
tag_freq = nltk.FreqDist(POS_tag_list)
#Sort the result 
sorted_tag_freq = sorted(tag_freq.items(), key = lambda k:k[1], reverse = True)
```

```python
sorted_tag_freq
```

```python
tag_freq.plot(30)
```

```python
#Generate a list of POS tags
POS_tag_list1 = [(word,tag) for (word,tag) in POS_tags if tag.startswith('N')&tag.endswith('P')]
```
